# Pandas 자료구조 (Series, DataFrame)

- Pandas의 Series는 1차원, DataFrame은 2차원 데이터를 다루는 자료구조
- DataFrames are the primary data structure in pandas, representing tabular data with rows and columns.
- 일반적으로 column은 변수, row는 샘플
- 리스트와 딕셔너리의 조합형
- 숫자형, 문자형, 범주형 등의 다양한 데이터 입력 가능

````python
from pandas import Series, DataFrame
````

````python
genes = Series([0.1, 0.2, 1.4, 0.6, 1.1])
print(genes)
````

````python
genes = Series([0.1, 0.2, 1.4, 0.6, 1.1], index=['A', 'B', 'C', 'D', 'E'])
print(genes)
````

#### Create from a dictionary

````python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Height': [5.5, 6.0, 5.8]
}

df = pd.DataFrame(data)
print(df)
````

#### Create from a list of dictionary

````python
data = [
    {'Name': 'Alice', 'Age': 25, 'Height': 5.5},
    {'Name': 'Bob', 'Age': 30, 'Height': 6.0},
    {'Name': 'Charlie', 'Age': 35, 'Height': 5.8}
]

df = pd.DataFrame(data)
print(df)
````

* 인덱스 자동 정렬, 행렬 연산

````python
genes1 = Series([0.1, 0.2, 1.4, 0.6, 1.1], index=['A', 'B', 'C', 'D', 'E'])
genes2 = Series([0.1, 0.2, 1.4, 0.6, 1.1], index=['B', 'C', 'D', 'E', 'A'])
genes1 + genes2
````

````python
print(genes2.sort_values())
print(genes2.sort_index())
````

* DataFrame 생성은 '{', '}' 이용
* DataFrame은 Series의 집합

````python
genes = {'A': [0.5, 0.1, 0.3],
         'B': [0.8, 0.9, 0.4]}
print(genes)
genes_df = DataFrame(genes)
print(genes_df)
print(genes_df['A'])
print(type(genes_df['A']))
````

````python
genes = {'A': [0.5, 0.1, 0.3],
         'B': [0.8, 0.9, 0.4]}
genes_df = DataFrame(genes, columns=['B', 'A'], index=['day1', 'day2', 'day3'])
print(genes_df)
````

````python
print(genes_df['A'])
print(genes_df.loc['day1'])
print(genes_df.index)
print(list(genes_df.columns))
````

### Example: Analyzing and Visualizing DNA Sequence Data

Let's combine these libraries in a practical example. We'll read a CSV file containing DNA sequences and their properties, perform some analysis, and visualize the results.

**Example Data (`dna_data.csv`):**
```csv
Sequence,Length,GC_Content
ATGCGTACGTAGCTAGCTAG,20,50.0
GCGTACGTAGCTAGCTAGCA,20,55.0
TGCATGCATGCTAGCTAGCA,20,45.0
```

**Python Script:**
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv('dna_data.csv')

# Calculate basic statistics
mean_gc = df['GC_Content'].mean()
std_gc = df['GC_Content'].std()

print(f"Mean GC Content: {mean_gc:.2f}%")
print(f"Standard Deviation of GC Content: {std_gc:.2f}%")

# Visualize the data
plt.figure(figsize=(10, 5))

# Histogram of GC content
plt.subplot(1, 2, 1)
sns.histplot(df['GC_Content'], kde=True)
plt.title('GC Content Distribution')
plt.xlabel('GC Content (%)')

# Boxplot of sequence lengths
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Length'])
plt.title('Sequence Length Distribution')
plt.xlabel('Length (bp)')

plt.tight_layout()
plt.show()
```

### Explanation:
- **Reading Data**: We read the DNA sequence data from a CSV file into a Pandas DataFrame.
- **Calculating Statistics**: We calculate the mean and standard deviation of the GC content using Pandas.
- **Visualizing Data**: We create a histogram of the GC content and a boxplot of the sequence lengths using Matplotlib and Seaborn.

This script demonstrates the integration of data handling, analysis, and visualization in Python. By learning and practicing these concepts, you'll be well-equipped to handle and analyze biological data.
